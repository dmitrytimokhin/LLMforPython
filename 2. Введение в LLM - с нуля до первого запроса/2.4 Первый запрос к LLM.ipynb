## Запрос к LLM через OpenRouter API (Альтернатива Mistral)

```python
pip install openai -q
```

#### Создайте файл first_openrouter.py:

```python
from openai import OpenAI

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key="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",
)

response = client.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=[
    {"role": "user", "content": "Привет! Напиши кратко, что такое FastAPI."}
    ],
    temperature=0.2,
    max_tokens=100
)

print(response.choices[0].message.content)
```

**Шаг 4. Запустите**

#### Советы для продакшена
1. Используйте модели с суффиксом :free - Только они действительно бесплатны. Модели без суффикса списывают кредиты (даже если баланс нулевой — запрос просто не пройдёт).
2. Всегда ограничивайте max_tokens
3. Обрабатывайте ошибки 

#### Ресурсы

* Каталог моделей (https://openrouter.ai/models) — фильтры по цене, контексту, языкам.
* OpenAI Python SDK (https://github.com/openai/openai-python) — официальная документация, включая работу с объектом ответа и типизацию.
* Responses API Guide (https://developers.openai.com/api/reference/responses/overview) — руководство по новому упрощённому интерфейсу инференса.
* OpenRouter API Reference (https://openrouter.ai/docs/api-reference/chat-completion) — спецификация эндпоинтов и форматов запросов/ответов.

```python
# Классический Chat Completions API
response = client.chat.completions.create(
    model="openai/gpt-oss-20b:free",
    messages=[{"role": "user", "content": "Привет!"}]
)
print(response.choices[0].message.content)
```

```python
# Новый Responses API (если поддерживается провайдером)
response = await client.responses.create(
    model="openai/gpt-oss-20b:free",
    input="Привет!"
)

print(response.output_text)
```
