### 1) Простой запрос

```python
import ollama

response = ollama.generate(
    model='mistral',
    prompt='Что такое Alembic?'
)
print(response['response'])
```

### 2) Чат-режим с поддержкой диалога

```python
response = ollama.chat(
    model='mistral',
    messages=[{'role': 'user', 'content': 'Что такое Alembic?'}],
    options={'temperature': 0.2, 'num_ctx': 2048}
)
print(response['message']['content'])
```

### 3) Потоковая генерация: («печатание» в реальном времени):

```python
for chunk in ollama.chat(
    model='mistral',
    messages=[{'role': 'user', 'content': 'Расскажи про миграции в БД'}],
    stream=True
):
    print(chunk['message']['content'], end='', flush=True)
```

---
* **num_ctx** — это размер контекстного окна в токенах. Он определяет, сколько токенов (запрос + история + ответ) модель может обработать за один раз.

**Рекомендации:** оставляйте 2048 для чата, ставьте 4096–8192 для RAG.

### 4) REST API через requests
**Важно:** сервер Ollama должен быть запущен. (ollama run mistral)

```python
import requests

response = requests.post(
    "http://localhost:11434/api/generate",
    json={
        "model": "mistral",
        "prompt": "Что такое Alembic?",
        "stream": False,
        "options": {
            "temperature": 0.2,
            "num_ctx": 2048
        }
    }
)

print(response.json()["response"])
```

Советы по выбору модели
---
| Модель | Параметры | Для чего подходит |
| :--- | :--- | :--- |
| `phi3` | `3.8B` | Самая лёгкая — работает даже на слабых ПК (CPU-only) |
| `mistral` | `7B` | Лучший баланс скорости, качества и потребления памяти |
| `llama3` | `8B` / `70B` | Универсальная модель, отлично подходит для RAG и агентов |
| `gemma4:e2b` | `2B` | Очень компактная, но уступает `phi3` в качестве |
| `qwen3:4b` | `4B` | Хорошее качество при умеренном размере |

---
| Сценарий использования              | Рекомендуемая модель |
|-------------------------------------|---------------------|
| 🖥️ Слабый ПК / ноутбук без GPU     | `phi3` или `gemma4:e2b` |
| ⚡ Быстрый прототип / локальный запуск | `mistral` или `qwen3:4b` |
| 🧠 Сложные задачи / RAG / агенты    | `llama3:70B` (если есть ресурсы) |
| 🔁 Баланс качества и скорости       | `llama3:8B` или `mistral` |
