### Ключевые production-паттерны

|Паттерн|Зачем|
|:---|:---|
|AsyncQdrantClient + AsyncClient|Не блокировать event loop → высокая пропускная способность|
|Pydantic Settings + .env|Безопасная конфигурация, easy override в Docker/K8s|
|Изоляция embedding / генерации|Чёткое разделение ролей → масштабируемость|

#### Примечание
В продакшене не пересоздавайте коллекцию на каждом запуске — уберите **delete_collection** и добавьте проверку версии данных.

Для деплоя используйте **Docker-compose.**
